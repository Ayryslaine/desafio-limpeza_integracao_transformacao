```python
df_avaliacoes = pd.read_json('/content/avaliacoes.json')

display(df_avaliacoes)
```

# Limpeza dos Banco de Dados

**Preencher valores nulos no arquivo CSV - clientes**

```python
df_clientes["Telefone"] = df_clientes["Telefone"].fillna("Desconhecido")
```

**Retirando valores Negativos da coluna Renda Mensal**

```python
df_clientes["Renda_Mensal"] = df_clientes["Renda_Mensal"].abs()
```

**Removendo outliers da coluna "Idade"**

```python
df_clientes = df_clientes[df_clientes["Idade"] <= 122]
```

# Padronização dos Banco de Dados

**Arquivo Json**

```python
#Padronizando data

if "Data_Avaliacao" in df_avaliacoes.columns:
    df_avaliacoes["Data_Avaliacao"] = pd.to_datetime(df_avaliacoes["Data_Avaliacao"], errors="coerce")

df_avaliacoes["Data_Avaliacao"] = df_avaliacoes["Data_Avaliacao"].dt.strftime('%d/%m/%Y')
```

```python
print("Coluna Recomenda antes:\n")
df_avaliacoes['Recomenda'].unique()
```

```python
#Padronizando Coluna "Recomenda"

df_avaliacoes['Recomenda'] = df_avaliacoes['Recomenda'].replace({
    False: 'Não',
    True: 'Sim',
}).astype(str).str.replace('false', 'Não').str.replace('true', 'Sim')

print("Coluna Recomenda depois:\n")
df_avaliacoes['Recomenda'].unique()
```

**Arquivo CSV - Clientes**

```python
#Padronizando data

# Padroniza separadores: troca "/" por "-"
df_clientes["Data_Cadastro"] = df_clientes["Data_Cadastro"].astype(str).str.replace("/", "-", regex=False)

# Converte para datetime (aceitando dia/mês/ano)
df_clientes["Data_Cadastro"] = pd.to_datetime(
    df_clientes["Data_Cadastro"],
    errors="coerce",
    dayfirst=True
)

# Substitui valores inválidos (NaT) por uma data padrão
df_clientes["Data_Cadastro"] = df_clientes["Data_Cadastro"].fillna(pd.to_datetime("2000-01-01"))

# Converte para string no formato brasileiro
df_clientes["Data_Cadastro"] = df_clientes["Data_Cadastro"].dt.strftime("%d/%m/%Y")

df_clientes
```

```python
print("Coluna Ativo antes:\n")
df_clientes["Ativo"].unique()
```

```python
#Padronizando Coluna "Ativo"

df_clientes["Ativo"] = df_clientes["Ativo"].replace({
    "No": 'Não',
    "Yes": 'Sim',
    "S": "Sim",
    "N": "Não"
})

print("Coluna Ativo depois:\n")
df_clientes["Ativo"].unique()
```

**Arquivo Excel**

```python
#Padronizando data

if "Data_Entrada_Estoque" in df_produtos.columns:
    df_produtos["Data_Entrada_Estoque"] = pd.to_datetime(df_produtos["Data_Entrada_Estoque"], errors="coerce")

df_produtos["Data_Entrada_Estoque"] = df_produtos["Data_Entrada_Estoque"].dt.strftime('%d/%m/%Y')
```

```python
df_produtos
```

**Arquivo CSV - Vendas**

```python
#Padronizando data

# Padroniza separadores: troca "/" por "-"
df_vendas["Data_Venda"] = df_vendas["Data_Venda"].astype(str).str.replace("/", "-", regex=False)

# Converte para datetime (aceitando dia/mês/ano)
df_vendas["Data_Venda"] = pd.to_datetime(
    df_vendas["Data_Venda"],
    errors="coerce",
    dayfirst=True
)

# Substitui valores inválidos (NaT) por uma data padrão
df_vendas["Data_Venda"] = df_vendas["Data_Venda"].fillna(pd.to_datetime("2000-01-01"))

# Converte para string no formato brasileiro
df_vendas["Data_Venda"] = df_vendas["Data_Venda"].dt.strftime("%d/%m/%Y")

df_vendas
```

# Integrando os Banco de Dados

```python
# Vendas + Clientes
df_integrado = df_vendas.merge(df_clientes, on="ID_Cliente", how="left")

# Vendas + Produtos (usando nome do produto)
df_integrado = df_integrado.merge(df_produtos, left_on="Produto", right_on="Nome_Produto", how="left")

# Vendas + Avaliações (cliente e produto avaliado)
df_integrado = df_integrado.merge(df_avaliacoes, left_on=["ID_Cliente", "Produto"], right_on=["ID_Cliente", "Produto_Avaliado"], how="left")

display(df_integrado)
```

# Limpeza Banco de Dados integrado

**Diagnóstico do banco de dados**

```python
def diagnostico(df, nome):
  print(f"\n--- Diagnóstico de {nome} ---\n")
  print(df_integrado.info())
  print("\nValores nulos por coluna:\n")
  display(df_integrado.isnull().sum())
  print("\nDuplicados:\n", df_integrado.duplicated().sum())

diagnostico(df_avaliacoes, "Banco de Dados integrado")
```

**colunas**

```python
df_integrado.columns
```

**Padronizando colunas para int**

```python
df_integrado["Idade"] = df_integrado["Idade"].fillna(0).astype(int)
df_integrado["Nota"] = df_integrado["Nota"].fillna(0).astype(int)
```

**Trocando valores nulos por "Desconhecido"**

```python
def nulo(df, coluna):
  df[coluna] = df[coluna].fillna("Desconhecido")

nulo(df_integrado, "Nome")
nulo(df_integrado, "Email")
nulo(df_integrado, "Telefone")
nulo(df_integrado, "Data_Cadastro")
nulo(df_integrado, "Cidade")
nulo(df_integrado, "Estado")
nulo(df_integrado, "Ativo")
nulo(df_integrado, "Avaliacao_ID")
nulo(df_integrado, "Produto_Avaliado")
nulo(df_integrado, "Comentario")
nulo(df_integrado, "Data_Avaliacao")
nulo(df_integrado, "Recomenda")
```

**Substituindo valores Nulos de colunas numéricas pela Média**

```python
def media(df,coluna):
  df[coluna] = df[coluna].fillna(df[coluna].mean())

media(df_integrado, "Idade")
media(df_integrado, "Renda_Mensal")
```

Grafico Linha = Renda Média Mensal ao Longo do Tempo de Cadastro

```python
df_integrado['Data_Cadastro'] = pd.to_datetime(df_integrado['Data_Cadastro'], format='%d/%m/%Y', errors='coerce')

income_over_time = df_integrado.groupby(df_integrado['Data_Cadastro'].dt.to_period('M'))['Renda_Mensal'].mean()
income_over_time.index = income_over_time.index.astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(x=income_over_time.index, y=income_over_time.values, marker='o', color='skyblue')
plt.title('Renda Média Mensal ao Longo do Tempo de Cadastro')
plt.xlabel('Mês de Cadastro')
plt.ylabel('Renda Média Mensal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
